# Kmeans-Clustering

```python
from sklearn.cluster import KMeans
```

```python
#let x be the feature set

x =df.iloc[:,3:12].values
x
```

```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters = i,n_init = 10,random_state = 101)
    km.fit(x)
    wcss.append(km.inertia_)
    
plt.figure(figsize = (8,4))  
plt.plot(range(1,11),wcss)
plt.title("Elbow Method")
plt.xlabel("No of clusters k")
plt.ylabel("WCSS score")
plt.show()
```

```python
#here optimum no of cluster is 3 since it is at the elbow point

km = KMeans(n_clusters = 3,n_init = 10,random_state = 101)
km.fit(x)

df['cluster_labels'] = km.labels_
df
```

```python
plt.figure(figsize = (10,7))                 
plt.scatter(x=df['Magnesium'],y=df['Proanthocyanins'],c=df['cluster_labels'])
plt.show()
```

```python
plt.figure(figsize = (10,7))                 
plt.scatter(x=df['Magnesium'],y=df['Alcohol'],c=df['cluster_labels'])
plt.show()
```

```python
plt.figure(figsize = (10,7))                 
plt.scatter(x=df['Magnesium'],y=df['Total_Phenols'],c=df['cluster_labels'])
plt.show()
```

# Hierarchical Clustering

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df.loc[:,:] = sc.fit_transform(df)
```

```python
mergings = linkage(df, method = "complete")
```

```python
plt.figure(figsize=(10,6))
dendrogram(mergings) 
plt.title('Dendrogram')
plt.show()
```

```python
mergings.shape
```

```python
from scipy.cluster.hierarchy import fcluster
labels = fcluster(mergings,8,criterion = "distance")
```

```python
pd.Series(labels).value_counts()
```

```python
#we got 6 clusters from the dendogram
```
